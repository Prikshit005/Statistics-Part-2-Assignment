**1. What is hypothesis testing in statistics?**
- Hypothesis testing in statistics is a method used to make decisions or draw conclusions about a population based on a sample of data. It starts with a hypothesis, which is a statement or assumption about a population parameter (like the average or proportion). There are usually two hypotheses: the null hypothesis (H₀), which assumes nothing has changed or there's no effect, and the alternative hypothesis (H₁ or Ha), which suggests there is a change or effect. Using sample data, statisticians perform a test to see how likely the observed results are if the null hypothesis were true. If the results are very unlikely under the null hypothesis, it is rejected in favor of the alternative. This process helps researchers make informed decisions based on evidence.

**2. What is the null hypothesis, and how does it differ from the alternative hypothesis?**
- The null hypothesis (H₀) is a statement that there is no effect, no difference, or nothing new happening—it represents the standard or accepted belief. In contrast, the alternative hypothesis (H₁ or Ha) is what we want to prove; it suggests that there is a real effect, a difference, or a change. For example, if we're testing a new medicine, the null hypothesis might say the medicine has no effect, while the alternative hypothesis says the medicine does help. In hypothesis testing, we assume the null is true until the evidence (data) shows it's very unlikely, at which point we may reject it in favor of the alternative.

**3. What is the significance level in hypothesis testing, and why is it important?**
- The significance level in hypothesis testing, usually written as alpha (α), is the threshold we set to decide whether to reject the null hypothesis. It represents the chance of making a false positive error—rejecting the null hypothesis when it is actually true. Common values are 0.05 or 0.01, meaning there's a 5% or 1% risk of being wrong if we reject the null. It's important because it helps control how much risk we're willing to take when making a conclusion based on data. A smaller significance level means we're being more cautious about claiming an effect or difference exists.

**4. What does a P-value represent in hypothesis testing?**
- The P-value in hypothesis testing tells us how likely it is to get the results we observed (or more extreme) if the null hypothesis is actually true. It's a number between 0 and 1. A small P-value (usually less than the significance level, like 0.05) means the results are unlikely under the null hypothesis, so we may reject it and support the alternative hypothesis. A large P-value means the results are more likely under the null, so we don't have enough evidence to reject it. In simple terms, the P-value helps us decide whether the data supports or goes against the null hypothesis.

**5. How do you interpret the P-value in hypothesis testing?**
- To interpret the P-value in hypothesis testing, we can compare it to our chosen significance level (like 0.05). If the P-value is less than or equal to the significance level, it means the results are unlikely to have happened by chance, so we can reject the null hypothesis and consider the alternative hypothesis more likely. If the P-value is greater than the significance level, it means the results are likely due to random chance, so we do not reject the null hypothesis. In simple terms, a low P-value suggests strong evidence against the null, while a high P-value means there's not enough evidence to say anything has changed.

**6. What are Type 1 and Type 2 errors in hypothesis testing**
- In hypothesis testing, a Type 1 error happens when we reject the null hypothesis when it is actually true—this means we think there is an effect or difference when there really isn’t (a false alarm). A Type 2 error occurs when we fail to reject the null hypothesis when it is actually false—this means we miss a real effect or difference (a missed detection). In simple terms, Type 1 error is like a false positive, and Type 2 error is like a false negative. Both are mistakes, and understanding them helps researchers balance the risks when making decisions based on data.

**7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**
- The difference between a one-tailed and a two-tailed test in hypothesis testing lies in the direction of the effect we're testing for. A one-tailed test checks for an effect in only one direction—for example, whether a new drug is better than the current one. A two-tailed test checks for an effect in both directions—whether the new drug is either better or worse than the current one. In simple terms, a one-tailed test asks “Is it greater (or less)?” while a two-tailed test asks “Is it different?” We can choose the type of test based on our research question.

**8. What is the Z-test, and when is it used in hypothesis testing?**
- The Z-test is a statistical test used in hypothesis testing to determine whether there is a significant difference between sample and population means, or between two sample means, when the population standard deviation is known and the sample size is large (usually over 30). It uses the Z-distribution, which is a standard normal distribution. The Z-test is often used in situations like testing whether a new teaching method changes average test scores compared to the national average. In simple terms, it's a way to check if the results we see are unusual or expected, based on what we know about the population.

**9. How do you calculate the Z-score, and what does it represent in hypothesis testing?**
- The Z-score is calculated using the formula:
Z = (X̄ – μ) / (σ / √n),
where X̄ is the sample mean, μ is the population mean, σ is the population standard deviation, and n is the sample size. The Z-score tells us how many standard deviations our sample mean is away from the population mean. In hypothesis testing, it helps us decide whether the sample result is close enough to the population mean (supporting the null hypothesis) or far enough away to consider it unusual (possibly supporting the alternative hypothesis). In simple terms, a high or low Z-score shows the result is rare if the null hypothesis is true.

**10. What is the T-distribution, and when should it be used instead of the normal distribution?**
- The T-distribution is a type of probability distribution that is similar to the normal distribution but has heavier tails, meaning it accounts for more variability in the data. It is used in hypothesis testing when the sample size is small (usually less than 30) and the population standard deviation is unknown. In these cases, we can’t rely on the normal (Z) distribution because the estimates are less precise. The T-distribution helps give more accurate results by adjusting for the extra uncertainty. In simple terms, use the T-distribution when we have a small sample and don’t know the population standard deviation.

**11. What is the difference between a Z-test and a T-test?**
- The main difference between a Z-test and a T-test is when and how they are used. A Z-test is used when the population standard deviation is known and the sample size is large (usually over 30). It uses the normal distribution to find the test statistic. A T-test is used when the population standard deviation is unknown and the sample size is small (usually less than 30). It uses the T-distribution, which adjusts for extra uncertainty in small samples. In simple terms, use a Z-test when we know more about the population and have a big sample, and use a T-test when we know less and have a smaller sample.

**12. What is the T-test, and how is it used in hypothesis testing?**
- The T-test is a statistical test used in hypothesis testing to find out if there is a significant difference between the means of two groups or between a sample mean and a population mean when the population standard deviation is unknown and the sample size is small. It helps determine whether any difference observed is likely due to chance or if it reflects a real effect. For example, we might use a T-test to check if students’ average test scores changed after a new teaching method. In simple terms, the T-test helps us decide if the data shows a meaningful difference when we don’t have complete information about the population.

**13. What is the relationship between Z-test and T-test in hypothesis testing?**
- The Z-test and T-test are both used in hypothesis testing to compare means, but they are closely related based on the information we have. When the population standard deviation is known and the sample size is large, the Z-test is used because it relies on the normal distribution. When the population standard deviation is unknown and the sample size is small, the T-test is used because it adjusts for extra uncertainty using the T-distribution. As the sample size gets bigger, the T-distribution becomes more like the normal distribution, so the T-test and Z-test give very similar results. In simple terms, the T-test is like a more flexible version of the Z-test used when less is known about the population.

**14. What is a confidence interval, and how is it used to interpret statistical results?**
- A confidence interval is a range of values calculated from sample data that is likely to contain the true population parameter (like a mean or proportion) with a certain level of confidence, usually 95%. For example, if we have a 95% confidence interval for an average height as 160 cm to 170 cm, it means we can be 95% confident the true average height in the whole population falls within that range. Confidence intervals help interpret statistical results by showing not just a single estimate, but a range that accounts for uncertainty, making it easier to understand how precise and reliable the estimate is.

**15. What is the margin of error, and how does it affect the confidence interval?**
- The margin of error is the amount added and subtracted from a sample estimate to create a confidence interval. It shows how much the estimate might vary due to random sampling. A larger margin of error means the confidence interval is wider, so there’s more uncertainty about the true population value. A smaller margin of error means a narrower confidence interval and more precise estimates. In simple terms, the margin of error controls the size of the confidence interval and reflects how confident we are that the interval includes the true value.

**16. How is Bayes' Theorem used in statistics, and what is its significance?**
- Bayes' Theorem is used in statistics to update the probability of an event based on new evidence or information. It helps combine what we already believe (called the prior probability) with new data to get a better, updated probability (called the posterior probability). For example, if we want to know the chance someone has a disease given a positive test result, Bayes’ Theorem helps calculate that by considering both the test accuracy and how common the disease is. Its significance lies in allowing more accurate and logical decision-making by continuously refining probabilities as new data comes in.

**17. What is the Chi-square distribution, and when is it used?**
- The Chi-square distribution is a special probability distribution used in statistics, especially for tests involving categories or counts, like checking if observed data fits an expected pattern. It is shaped differently from the normal distribution and depends on something called "degrees of freedom." The Chi-square test is often used to see if there’s a relationship between two categorical variables (like gender and voting preference) or to test if a sample matches a known distribution. In simple terms, it helps answer questions about whether differences in categories are due to chance or something meaningful.

**18. What is the Chi-square goodness of fit test, and how is it applied?**
- The Chi-square goodness of fit test is a way to see if observed data matches what we would expect based on a specific theory or distribution. For example, if we roll a die many times, this test can check whether the results fit the idea that the die is fair (each number appears equally). We can compare the actual counts with the expected counts and use the Chi-square formula to see if the differences are big enough to be unlikely by chance. If they are, we might conclude the data doesn’t fit the expected pattern. In simple terms, it helps test if our observed results match what we would normally expect.

**19. What is the F-distribution, and when is it used in hypothesis testing?**
- The F-distribution is a probability distribution that is used mainly to compare two variances (how spread out data are) and to test if they are significantly different. It looks different from the normal distribution and depends on two sets of degrees of freedom. In hypothesis testing, the F-distribution is often used in ANOVA (Analysis of Variance) tests to check if the means of three or more groups are different by comparing the variation between groups to the variation within groups. In simple terms, it helps decide if differences in data groups are real or just happened by chance.

**20. What is an ANOVA test, and what are its assumptions?**
- An ANOVA test is a way to check if there are differences in the average values between three or more groups. Instead of comparing groups one by one, it looks at all of them together to see if at least one group is different from the rest. For the test to work well, it assumes that the groups are independent, the data in each group is normally distributed, and the groups have similar amounts of variation. In simple terms, ANOVA helps find out if group differences are real or just happened by chance, as long as these basic rules are followed.

**21. What are the different types of ANOVA tests?**
- There are several types of ANOVA tests, each used depending on the study design and the number of factors involved:

 - **One-way ANOVA:** Tests differences between the means of three or more groups based on one factor or variable (like comparing test scores across different schools).

 - **Two-way ANOVA:** Looks at the effects of two different factors at the same time (like testing how both teaching method and gender affect test scores), and can also check if the two factors interact.

 - **Repeated Measures ANOVA:** Used when the same subjects are tested multiple times under different conditions (like measuring blood pressure before and after taking medicine), to see if there are differences over time or conditions.

 - **Mixed-Design ANOVA:** Combines both between-group factors (different groups) and within-group factors (same subjects measured multiple times).

 In simple terms, the type of ANOVA we can use depends on how many factors y]we’re studying and whether we measure the same people more than once.

**22. What is the F-test, and how does it relate to hypothesis testing?**
- The F-test is a statistical test used in hypothesis testing to compare two or more group variances or means and see if they are significantly different. It works by calculating an F-statistic, which is a ratio of two types of variation—like the variation between group means compared to the variation within the groups. If this ratio is large, it suggests that the groups are different enough to reject the null hypothesis (which says all groups are the same). The F-test is commonly used in ANOVA tests to check if there are differences among multiple groups. In simple terms, the F-test helps decide if the differences in our data are real or just due to random chance.

**3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.**

```python
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x (int): The outcome (0 or 1)
    p (float): The probability of success (1), between 0 and 1

    Returns:
    float: Probability of outcome x
    """
    if x not in [0, 1]:
        return 0
    return p**x * (1 - p)**(1 - x)

p = 0.7
print(f"P(X=1) = {bernoulli_pdf(1, p)}")
print(f"P(X=0) = {bernoulli_pdf(0, p)}")
```

**9. Write a Python function to calculate Z-scores from a dataset and plot them.**

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_z_scores(data):
    """
    Calculate Z-scores for the given dataset and plot them.

    Parameters:
    data (array-like): List or array of numerical data points
    """
    data = np.array(data)
    mean = np.mean(data)
    std_dev = np.std(data)


    z_scores = (data - mean) / std_dev


    plt.scatter(range(len(data)), z_scores, color='blue')
    plt.axhline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.axhline(1, color='green', linestyle='--', label='Z=1')
    plt.axhline(-1, color='green', linestyle='--', label='Z=-1')
    plt.title('Z-scores of Data Points')
    plt.xlabel('Data Point Index')
    plt.ylabel('Z-score')
    plt.legend()
    plt.grid(True)
    plt.show()


sample_data = [10, 12, 9, 15, 14, 8, 7, 13, 11, 9]
plot_z_scores(sample_data)
```

**14.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_and_plot_z_score(data_point, data):
    """
    Calculate Z-score of data_point based on the data sample,
    and plot it on the standard normal distribution.

    Parameters:
    data_point (float): The data point to calculate Z-score for
    data (list or array): The dataset to compute mean and std deviation
    """
    mean = np.mean(data)
    std_dev = np.std(data)


    z = (data_point - mean) / std_dev
    print(f"Data point: {data_point}")
    print(f"Mean of data: {mean:.2f}")
    print(f"Standard deviation of data: {std_dev:.2f}")
    print(f"Z-score: {z:.2f}")


    x = np.linspace(-4, 4, 1000)
    plt.plot(x, norm.pdf(x), label='Standard Normal Distribution')

    plt.axvline(z, color='red', linestyle='--', label=f'Z-score = {z:.2f}')
    plt.fill_between(x, 0, norm.pdf(x), where=(x <= z), color='red', alpha=0.3)

    plt.title('Z-score on Standard Normal Distribution')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

sample_data = [10, 12, 9, 15, 14, 8, 7, 13, 11, 9]
point = 14

calculate_and_plot_z_score(point, sample_data)
```

**15. Implement hypothesis testing using Z-statistics for a sample dataset.**

```python
import numpy as np
from scipy.stats import norm

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Perform Z-test for the sample mean against the population mean.

    Parameters:
    sample (list or array): Sample data
    pop_mean (float): Population mean under the null hypothesis
    pop_std (float): Population standard deviation (known)
    alpha (float): Significance level (default 0.05)

    Returns:
    dict: Test statistic, p-value, and conclusion
    """
    n = len(sample)
    sample_mean = np.mean(sample)


    std_error = pop_std / np.sqrt(n)


    z_stat = (sample_mean - pop_mean) / std_error

    p_value = 2 * (1 - norm.cdf(abs(z_stat)))


    reject_null = p_value < alpha
    conclusion = "Reject the null hypothesis." if reject_null else "Fail to reject the null hypothesis."

    return {
        'z_statistic': z_stat,
        'p_value': p_value,
        'conclusion': conclusion
    }

sample_data = [102, 100, 98, 101, 99, 97, 103, 100, 102, 98]
population_mean = 100
population_std = 5

result = z_test(sample_data, population_mean, population_std)

print(f"Z-statistic: {result['z_statistic']:.3f}")
print(f"P-value: {result['p_value']:.4f}")
print(f"Conclusion: {result['conclusion']}")
```

**16. Create a confidence interval for a dataset using Python and interpret the result.**

```python
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of the data.

    Parameters:
    data (list or array): Sample data
    confidence (float): Confidence level (default 0.95 for 95% CI)

    Returns:
    tuple: (lower bound, upper bound)
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)


    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_crit * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

sample_data = [12, 15, 14, 10, 13, 14, 16, 12, 11, 15]

lower, upper = confidence_interval(sample_data, confidence=0.95)
mean = np.mean(sample_data)

print(f"Sample Mean = {mean:.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
print(f"Interpretation: We are 95% confident that the true population mean lies between {lower:.2f} and {upper:.2f}.")
```

**1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.**

```python
import numpy as np
from scipy.stats import norm

def z_test(sample, pop_mean, pop_std, alpha=0.05):

    n = len(sample)
    sample_mean = np.mean(sample)


    std_error = pop_std / np.sqrt(n)


    z_stat = (sample_mean - pop_mean) / std_error


    p_value = 2 * (1 - norm.cdf(abs(z_stat)))


    if p_value < alpha:
        conclusion = "Reject the null hypothesis: sample mean is significantly different from population mean."
    else:
        conclusion = "Fail to reject the null hypothesis: no significant difference detected."

    return {
        'z_statistic': z_stat,
        'p_value': p_value,
        'conclusion': conclusion
    }


sample_data = [102, 100, 98, 101, 99, 97, 103, 100, 102, 98]
population_mean = 100
population_std = 5

result = z_test(sample_data, population_mean, population_std)

print(f"Z-statistic: {result['z_statistic']:.3f}")
print(f"P-value: {result['p_value']:.4f}")
print(f"Conclusion: {result['conclusion']}")
```

**2. Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python.**

```python
import numpy as np
from scipy.stats import norm

def two_sample_z_test(sample1, sample2, pop_std1, pop_std2, alpha=0.05):

    n1, n2 = len(sample1), len(sample2)
    mean1, mean2 = np.mean(sample1), np.mean(sample2)

    se_diff = np.sqrt((pop_std1**2 / n1) + (pop_std2**2 / n2))

    z_stat = (mean1 - mean2) / se_diff


    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    conclusion = "Reject null hypothesis: means are significantly different." if p_value < alpha else "Fail to reject null hypothesis: no significant difference."

    return {'z_statistic': z_stat, 'p_value': p_value, 'conclusion': conclusion}

np.random.seed(123)
pop_mean1, pop_std1 = 50, 5
pop_mean2, pop_std2 = 52, 5

sample1 = np.random.normal(loc=pop_mean1, scale=pop_std1, size=30)
sample2 = np.random.normal(loc=pop_mean2, scale=pop_std2, size=30)

result = two_sample_z_test(sample1, sample2, pop_std1, pop_std2)

print(f"Sample 1 mean: {np.mean(sample1):.2f}")
print(f"Sample 2 mean: {np.mean(sample2):.2f}")
print(f"Z-statistic: {result['z_statistic']:.3f}")
print(f"P-value: {result['p_value']:.4f}")
print(f"Conclusion: {result['conclusion']}")
```

**4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    std_error = pop_std / np.sqrt(n)
    z_stat = (sample_mean - pop_mean) / std_error
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    reject = p_value < alpha
    conclusion = "Reject null hypothesis" if reject else "Fail to reject null hypothesis"

    return z_stat, p_value, conclusion

def plot_z_test(z_stat, alpha=0.05):

    z_critical = norm.ppf(1 - alpha/2)

    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.figure(figsize=(10,6))
    plt.plot(x, y, label='Standard Normal Distribution')


    plt.fill_between(x, 0, y, where=(x <= -z_critical), color='red', alpha=0.3, label='Rejection Region')
    plt.fill_between(x, 0, y, where=(x >= z_critical), color='red', alpha=0.3)


    plt.axvline(z_stat, color='blue', linestyle='--', label=f'Observed Z = {z_stat:.2f}')


    plt.axvline(-z_critical, color='red', linestyle='--', label=f'Critical Values ±{z_critical:.2f}')
    plt.axvline(z_critical, color='red', linestyle='--')

    plt.title('Two-Tailed Z-Test with Decision Regions')
    plt.xlabel('Z value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

np.random.seed(101)
sample = np.random.normal(loc=102, scale=5, size=40)
population_mean = 100
population_std = 5
alpha = 0.05

z_stat, p_value, conclusion = two_tailed_z_test(sample, population_mean, population_std, alpha)

print(f"Sample Mean: {np.mean(sample):.2f}")
print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_value:.4f}")
print(f"Conclusion: {conclusion}")


plot_z_test(z_stat, alpha)

```

**5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2(mu0, mu1, sigma, n, alpha=0.05):
  se = sigma / np.sqrt(n)
  z_critical = norm.ppf(1 - alpha)
    x_critical = mu0 + z_critical * se


    z_beta = (x_critical - mu1) / se
    beta = norm.cdf(z_beta)

    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)


    y0 = norm.pdf(x, mu0, se)
    y1 = norm.pdf(x, mu1, se)

    plt.figure(figsize=(10,6))
    plt.plot(x, y0, label="H₀: μ = {:.1f}".format(mu0), color="blue")
    plt.plot(x, y1, label="H₁: μ = {:.1f}".format(mu1), color="green")


    plt.fill_between(x, 0, y0, where=(x > x_critical), color='red', alpha=0.4, label="Type I Error (α)")


    plt.fill_between(x, 0, y1, where=(x < x_critical), color='orange', alpha=0.4, label="Type II Error (β)")


    plt.axvline(x_critical, color='black', linestyle='--', label=f'Critical value: {x_critical:.2f}')

    plt.title('Visualization of Type I and Type II Errors')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Output numerical results
    print(f"Critical value (x): {x_critical:.2f}")
    print(f"Type I Error (α): {alpha:.2f}")
    print(f"Type II Error (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {1 - beta:.4f}")

# Example usage:
visualize_type1_type2(mu0=100, mu1=103, sigma=10, n=30, alpha=0.05)
```

**9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.**

```python
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):

    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)

    t_crit = t.ppf((1 + confidence) / 2, df=n - 1)

    margin_of_error = t_crit * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    print(f"Sample Mean: {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
    print("\nInterpretation: We are", int(confidence*100),
          "% confident that the true population mean lies within this interval.")

    return lower_bound, upper_bound

np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)
confidence_interval(sample_data, confidence=0.95)
```

**10. Write a Python program to calculate the margin of error for a given confidence level using sample data.**

```python
import numpy as np
from scipy.stats import t

def calculate_margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for the sample mean at a given confidence level.

    Parameters:
    - data: array-like, sample data
    - confidence: float, confidence level (default 0.95)

    Returns:
    - margin_of_error: float
    """
    n = len(data)
    sample_std = np.std(data, ddof=1)
    std_error = sample_std / np.sqrt(n)


    t_critical = t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_critical * std_error

    print(f"Sample Size (n): {n}")
    print(f"Sample Standard Deviation: {sample_std:.2f}")
    print(f"Standard Error: {std_error:.2f}")
    print(f"T-critical value: {t_critical:.3f}")
    print(f"Margin of Error at {int(confidence*100)}% confidence: ±{margin_of_error:.2f}")

    return margin_of_error

np.random.seed(123)
sample_data = np.random.normal(loc=100, scale=12, size=40)
calculate_margin_of_error(sample_data, confidence=0.95)
```

**11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.**

```python
def bayes_theorem(prior_A, prob_B_given_A, prob_B_given_notA):

    prior_notA = 1 - prior_A
    prob_B = prob_B_given_A * prior_A + prob_B_given_notA * prior_notA
    posterior = (prob_B_given_A * prior_A) / prob_B
    return posterior

prior_disease = 0.01
prob_pos_given_disease = 0.99
prob_pos_given_no_disease = 0.05

posterior = bayes_theorem(prior_disease, prob_pos_given_disease, prob_pos_given_no_disease)

print(f"Probability of disease given a positive test: {posterior:.4f} ({posterior*100:.2f}%)")
```

**12. Perform a Chi-square test for independence between two categorical variables in Python.**
